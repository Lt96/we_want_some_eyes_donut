```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.segmentation import active_contour

img = cv2.imread("RAW_FUNDUS_INPUT/267_032.jpg")
row, column = np.shape(img[:,:,0])

original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

img_denoise = cv2.fastNlMeansDenoising(hsv[:,:,2],None,7,21)
cv2.namedWindow("ROI Selector",cv2.WINDOW_NORMAL)
ROI = cv2.selectROI("ROI Selector",img_denoise)
cv2.resizeWindow("ROI Selector" ,800,600)
cv2.waitKey(0)
cv2.destroyAllWindows()
img = img_denoise[int(ROI[1]):int(ROI[1]+ROI[3]), int(ROI[0]):int(ROI[0]+ROI[2])]
original = original[int(ROI[1]):int(ROI[1]+ROI[3]), int(ROI[0]):int(ROI[0]+ROI[2])]
s = np.linspace(0, 2*np.pi, 400)
r = int(img.shape[0]/2)
c = int(img.shape[1]/2)
#radius = int(input("Enter radius : "))
radius = 250
r = r + radius*np.sin(s) 
c = c + radius*np.cos(s)
init = np.array([r, c]).T

snake = active_contour(img, init, alpha=0.025, beta=70, w_line=0.125, w_edge=6.5, gamma=0.001,boundary_condition="periodic", 
                        max_px_move=1.0, max_iterations=2000, convergence=0.5, coordinates='rc')

fig, ax = plt.subplots(1,2,figsize=(7, 7))
ax[0].imshow(original)
ax[0].plot(init[:, 1], init[:, 0], '--r', lw=3)
ax[0].plot(snake[:, 1], snake[:, 0], '-k', lw=1)
ax[0].set_xticks([]), ax[0].set_yticks([])
ax[0].axis([0, original.shape[1], original.shape[0], 0])

ax[1].imshow(img)
ax[1].plot(init[:, 1], init[:, 0], '--r', lw=3)
ax[1].plot(snake[:, 1], snake[:, 0], '-k', lw=1)
ax[1].set_xticks([]), ax[1].set_yticks([])
ax[1].axis([0, img.shape[1], img.shape[0], 0])
```
